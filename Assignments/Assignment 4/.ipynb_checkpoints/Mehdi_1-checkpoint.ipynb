## Part a: Tone Generating 

This code cell is the answer to part a and serves the purpose of generating the tone associated with any key of the keypad:

### Instructions and Procedure of the Code

#### Function: `gentone`

This function generates the tone for a specified key on a telephone keypad.

**Parameters:**
- `key` (str): The key on the keypad (0-9, *, #). You can change it to any valid number.
- `duration` (float): Duration of the tone in seconds. Default is 1.0 second. This duration is chosen to provide a sufficiently long sample of the tone for listening and analysis. One second is typically long enough to recognize and analyze the tone but short enough to prevent excessively large data arrays.
- `sampling_rate` (int): The number of samples per second. Default is 8000 Hz. This is a standard sampling rate for telephone audio quality. It's high enough to accurately represent the frequencies used in the DTMF system (which are all below 1500 Hz), following the Nyquist theorem.

**Procedure:**
1. **Define Frequencies**: Two lists, `row_frequencies` and `col_frequencies`, are created to store the frequencies corresponding to the rows and columns of the keypad.

2. **Keypad Layout**: A 2D list `keypad` represents the layout of the keys.

3. **Find Key Position**: The function searches for the key in the `keypad` list to find its row and column position. If the key is not in the keypad, the function will inform the user to choose a key from the keypad.

4. **Determine Frequencies**: Using the row and column indices, it retrieves the corresponding frequencies from `row_frequencies` and `col_frequencies`.

5. **Generate Time Array**: A time array `t` is created using `np.linspace`, which spans from 0 to the specified `duration`. The number of points in this array is determined by `duration * sampling_rate` which are set as default to be 1.0 second and 8000 Hz respectively.

6. **Generate Tone**: The tone is generated as the average of two sine waves, each with a frequency corresponding to the row and column of the key.

7. **Return Tone**: The function returns the generated tone array.

#### Plotting

The script then uses the `gentone` function to generate a tone for a specific key (e.g., '5') and plots this tone as a function of time using `matplotlib`.

```python
# Part a: Tone Generating

def gentone(key, duration, sampling_rate):

    # Frequencies for each row and column of the keypad
    row_frequencies = [697, 770, 852, 941]
    col_frequencies = [1209, 1336, 1477]

    # Keypad layout
    keypad = [
        ['1', '2', '3'],
        ['4', '5', '6'],
        ['7', '8', '9'],
        ['*', '0', '#']
    ]

    # Check if the key is valid
    valid_key = False
    for row in keypad:
        if key in row:
            valid_key = True
            i, j = row.index(key), divmod(row.index(key), 3)[1]
            break

    if not valid_key:
        print(f"Invalid key: '{key}'. Please choose a key from the keypad.")
        return None

    # Frequencies corresponding to the key
    freq_row = row_frequencies[i]
    freq_col = col_frequencies[j]

    # Time array
    t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)

    # Generating the tone
    tone = (np.sin(2 * np.pi * freq_row * t) + np.sin(2 * np.pi * freq_col * t)) / 2

    return tone, freq_row, freq_col



# set duration and sampling_rate for generating the key
duration = 1.0  # Duration of the tone in seconds
sampling_rate = 8000  # Number of samples per second

# Generate the tone for any key
key = '5'  # set key from the keypad (0-9, *, #). Change it to any valid input.
tone, freq_row, freq_col = gentone(key, duration, sampling_rate)
time = np.linspace(0, duration, sampling_rate)

# print base frequencies for key
print(f"Base Frequencies for key {key} are {freq_row} and {freq_col}")

# Plotting
plt.figure(figsize=(10, 4))
plt.plot(time, tone)
plt.title(f"Tone for Key {key}")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

## Part b: Finding Frequencies Associated With the Input Signal

This code cell is the answer to part b and serves the purpose of finding frequencies associated with the signal for input key:

#### Fast Fourier Transform (FFT) Application

The `apply_fft` function is designed to analyze a time-domain signal and convert it into its frequency-domain representation using the Fast Fourier Transform (FFT). 

**Parameters:**
- `signal` (np.ndarray): The input signal in the time domain.
- `sampling_rate` (int): The rate at which the signal was sampled.

**Procedure:**
1. **Compute FFT**: The function calculates the FFT of the input signal using `np.fft.fft`, which transforms the signal from the time domain to the frequency domain.
2. **Frequency Array**: `np.fft.fftfreq` is used to create an array of frequencies corresponding to the FFT output. It takes the number of samples `N` and the inverse of the sampling rate as inputs.
3. **Magnitude Calculation**: The magnitude of the FFT output is computed using `np.abs`, as FFT returns complex numbers. This magnitude represents the signal's power spectrum.

#### Peak Reporting

The `report_peaks` function identifies and reports the significant peaks in the frequency spectrum.

**Parameters:**
- `frequencies` (np.ndarray): Array of frequencies from the FFT output.
- `magnitudes` (np.ndarray): Array of magnitudes from the FFT output.
- `threshold` (float): A threshold for peak detection. This is the smallest value that can be considered a peak in the magnitudes.

**Procedure:**
1. **Peak Detection**: Using `scipy.signal.find_peaks`, the function identifies the peaks in the magnitude array that are above a specified threshold.
2. **Extract Peak Information**: The frequencies and magnitudes of these peaks are extracted.
3. **Return Peaks**: The function returns a list of tuples, each containing the frequency and magnitude of a peak.

#### Frequency Identification and Plotting

- **Identify Relevant Frequencies**: The script filters the reported peaks to include only those within the relevant frequency range (0 to 1500 Hz in this case).
- **Print Frequencies**: It prints the frequencies associated with the key (e.g., '5'), which should correspond to the row and column frequencies of the keypad.
- **Plot Power Spectrum**: Finally, the power spectrum of the signal is plotted, showing the magnitudes of the frequency components. This plot is helpful in visually confirming the presence of the expected frequencies.

#### Execution Flow

1. The `apply_fft` function is applied to the signal generated for valid input key.
2. The `report_peaks` function then identifies the significant peaks in the frequency spectrum.
3. The script prints the relevant frequencies and plots the power spectrum, allowing for visual verification of the correct frequencies being detected.

This analysis confirms the accuracy frequencies associated with the signal key (e.g., '5') extracted form power spectrum plot with the base frequencies for the key (e.g., '5') calculated in part a.

```python
# Part b: Finding Frequencies Associated With the Input Signal

# Finction to analyze a time-domain signal and convert it into its frequency-domain representation using the Fast Fourier Transform (FFT). 
def apply_fft(signal, sampling_rate):
    # Number of samples
    N = len(signal)

    # FFT
    fft_output = np.fft.fft(signal)

    # Frequencies corresponding to the FFT output
    freq = np.fft.fftfreq(N, 1/sampling_rate)

    # Magnitude of the FFT (power spectrum)
    magnitude = np.abs(fft_output)

    return freq, magnitude


# Apply FFT to the Tone signal from part a
freq, magnitude = apply_fft(tone, sampling_rate)


# Function to find and report peaks in the frequency spectrum
def report_peaks(frequencies, magnitudes, threshold=1000):
    # Detect peaks in the magnitude spectrum
    # 'find_peaks' returns the indices of peaks that are higher than the specified threshold
    peaks, _ = find_peaks(magnitudes, height=threshold)

    # Extract the frequencies of the detected peaks
    # The indices from 'find_peaks' are used to retrieve corresponding frequencies
    peak_frequencies = frequencies[peaks]

    # Extract the magnitudes of the detected peaks
    # Similar to frequencies, the magnitudes at the peak indices are retrieved
    peak_magnitudes = magnitudes[peaks]

    # Combine frequencies and magnitudes into a list of tuples for each peak
    return list(zip(peak_frequencies, peak_magnitudes))

# Finding and reporting peaks for the signal of key 
peaks = report_peaks(freq, magnitude)

# Filtering out negative frequencies and showing relevant peaks
relevant_peaks = [(f, m) for f, m in peaks if 0 <= f <= 1500]

# print associated frequencies
print(f"Frequencies associated with the signal for key {key} extracted from the plot are {relevant_peaks[0][0]} Hz and {relevant_peaks[1][0]} Hz")

# Plotting the power spectrum
plt.figure(figsize=(10, 4))
plt.plot(freq, magnitude)
plt.title(f"Power Spectrum of the Signal for Key {key}")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.grid(True)
plt.xlim([0, 2000])  # Limiting x-axis to view relevant frequencies
plt.show()
```

## Part c: Generating a 2-Digit Time Signal with Noise

This code cell is the answer to part c and serves the purpose of generating a 2-digit time signal with some random noise:


#### Function: `generate_combined_signal`

This function creates a time-domain signal simulating the pressing of two different keys on a telephone keypad, including a pause between them and adding random noise.

**Parameters:**
- `key1`, `key2` (str): Keys on the keypad (0-9, *, #).
- `pause_duration` (float): Duration of the pause between the key presses.
- `duration` (float): Duration of the tone for each key press.
- `sampling_rate` (int): Sampling rate for the tone generation.

**Procedure:**
1. **Tone Generation**: The function uses `gentone` to generate the tone for each specified key (`key1` and `key2`), each lasting for the specified `duration`.

2. **Pause Creation**: A pause is generated as an array of zeros, with a length determined by the `pause_duration` and `sampling_rate`.

3. **Signal Combination**: The tones for `key1` and `key2` are concatenated with the pause array in between, forming a single combined signal.

4. **Noise Addition**: Random noise is generated using `np.random.normal` with a mean of 0 and a standard deviation of 0.01, ensuring the noise has a low amplitude. This noise is then added to the combined signal to simulate real-world signal imperfections.

5. **Return Noisy Signal**: The function returns the combined signal with added noise.

#### Plotting the Combined Signal

- **Time Array Creation**: A time array is created corresponding to the length of the combined signal, calculated based on the `sampling_rate`.

- **Signal Plotting**: The combined signal is plotted against time, displaying the tones for `key1` and `key2` with a pause in between, and the superimposed noise.

#### Execution Flow

1. The `generate_combined_signal` function is called with specified keys (e.g., '5' and '9'), `pause_duration`, `duration`, and `sampling_rate`.

2. The combined signal with noise is generated and stored in `combined_signal`.

3. The signal is plotted, showing the distinct segments for each key press and the pause, along with the noise pattern throughout the signal.

This implementation successfully creates a realistic simulation of pressing two keys on a keypad, which is crucial for analyzing how DTMF tones behave in real-world scenarios, especially with the presence of noise.

```python
# Part c: Generating a 2-Digit Time Signal with Noise

def generate_combined_signal(key1, key2, pause_duration, duration, sampling_rate):
    # Generate tones for each key
    tone1, freq_row1, freq_col1 = gentone(key1, duration, sampling_rate)
    tone2, freq_row2, freq_col2 = gentone(key2, duration, sampling_rate)

    # Generate a pause (silence) between the tones
    pause = np.zeros(int(pause_duration * sampling_rate))

    # Combine tones with the pause
    combined_signal = np.concatenate((tone1, pause, tone2))

    # Add random noise
    noise = np.random.normal(0, 0.01, combined_signal.shape)  # Low amplitude noise
    noisy_signal = combined_signal + noise

    return noisy_signal, freq_row1, freq_col1, freq_row2, freq_col2


# set pause_duration for pause between pressing two keys. Set duration and sampling_rate for generating each key.
pause_duration = 0.2
duration = 1.0
sampling_rate = 8000

# Generate a combined signal for pressing two keys with some random noise and pause duration
key1 = '5'  # set key from the keypad (0-9, *, #). Change it to any valid input.
key2 = '9'  # set key from the keypad (0-9, *, #). Change it to any valid input.

combined_signal, freq_row_key1, freq_col_key1, freq_col_key2, freq_col_key2 = generate_combined_signal(key1, key2, pause_duration, duration, sampling_rate)

# Create a time array for plotting
time = np.linspace(0, len(combined_signal) / sampling_rate, len(combined_signal))

# Plotting the combined signal
plt.figure(figsize=(12, 5))
plt.plot(time, combined_signal)
plt.title(f"Combined Signal for Keys {key1} and {key2} with Noise")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

## Part d: Applying a short-time Fourier transform (STFT) to combined signal

This code cell is the answer to part d and serves the purpose of applying a short-time Fourier transform (STFT) to combined signal in part c to extract the base frequincies for keys in combined signal:


#### Function: `apply_stft`

This function computes the Short-Time Fourier Transform (STFT) of a given signal. STFT is a method to analyze the varying frequency content of non-stationary signals.

**Parameters:**
- `signal` (np.ndarray): The input signal.
- `sampling_rate` (int): The sampling rate of the signal.
- `window` (str): The type of window to use in the STFT. Default is 'hann'. The Hann window, also known as the Hanning window, is a type of windowing function that is widely used in signal processing to reduce the spectral leakage. When you apply a Fourier Transform to a signal, you typically assume that the signal is periodic. However, when you take a short segment of a non-periodic signal, it creates discontinuities at the segment boundaries, leading to spectral leakage. The Hann window helps mitigate this effect by tapering the signal to zero at the segment boundaries, ensuring smoother transitions. This windowing creates a trade-off between frequency resolution and time resolution in the STFT.
- `nperseg` (int): The number of samples per segment for the STFT. In STFT, the signal is divided into overlapping segments, and the Fourier Transform is applied to each segment. The `nperseg` value, therefore, determines the size of these segments. A larger `nperseg` provides better frequency resolution (i.e., more detailed frequency information) but worse time resolution (i.e., less detailed time information). Conversely, a smaller `nperseg` provides better time resolution but worse frequency resolution. The value of 256 is a balance between these two aspects, suitable for many general applications, but this can be adjusted based on the specific requirements of the analysis.


**Procedure:**
1. **STFT Calculation**: The function uses `scipy.signal.stft` to compute the STFT of the signal. This involves segmenting the signal, applying the Hann window, and then performing the FFT on each segment.
2. **Magnitude Calculation**: It computes the magnitude of the STFT using `np.abs` to convert the complex STFT output to its magnitude, which is more interpretable.
3. **Return Values**: The function returns the time and frequency bins (`times`, `freqs`) and the magnitude of the STFT.

#### Base Frequency Printing

- **Base Frequencies**: The code prints the base frequencies for the keys '(e.g., '5' and '9') used in part (c). These frequencies are essential for comparing and verifying the STFT results.

#### Plotting the STFT

- **STFT Visualization**: The STFT is visualized using `plt.pcolormesh`, which creates a 2D colored plot. The x-axis represents time, the y-axis represents frequency, and the color intensity represents the magnitude of the STFT.
- **Frequency Range**: The frequency range is limited to 0-2000 Hz for clarity, encompassing the DTMF frequencies.
- **Title and Labels**: The plot is titled and labeled appropriately, indicating that it represents the STFT of the combined signal for keys (e.g., '5' and '9').

This analysis confirms the accuracy of frequencies associated with the signal keys (e.g., '5' and '9') extracted form STFT of the Combined Signal plot which are the brighter lines with the base frequencies for the keys (e.g., '5' and '9') calculated in part c.


```python
# Part d: Applying a short-time Fourier transform (STFT) to combined signal

# Applying STFT to the combined signal
def apply_stft(signal, sampling_rate, window='hann', nperseg=256):
    # Compute the STFT
    freqs, times, Zxx = stft(signal, fs=sampling_rate, window=window, nperseg=nperseg)
    
    # Calculate the magnitude of the STFT
    magnitude = np.abs(Zxx)
    
    # Return the time bins, frequency bins, and magnitude of the STFT
    return times, freqs, magnitude

# Generate the STFT for the combined signal
times, freqs, magnitude = apply_stft(combined_signal, sampling_rate)

# Printing base frequencies for the keys used in generating the combined signal
print("Base frequencies for keys in 2-digit time signal for comparison with STFT plot:")  
print(f"Base frequencies for first key {key1} are {freq_row_key1} and {freq_col_key1}")
print(f"Base frequencies for second key {key2} are {freq_row_key2} and {freq_col_key2}")


# Plotting the STFT
plt.figure(figsize=(12, 6))
plt.pcolormesh(times, freqs, magnitude, shading='gouraud') # Colored mesh plot for STFT
plt.title(f"STFT of the Combined Signal for Keys {key1} and {key2}")
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [s]')
plt.colorbar(label='Magnitude') # Color bar indicating the magnitude
plt.ylim([0, 2000])  # Limiting frequency range for clarity
plt.show()
```
