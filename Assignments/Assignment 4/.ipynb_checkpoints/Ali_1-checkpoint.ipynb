# Preparations

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Section (a)

```python
def gentone(num:int=0, duration:float=1e-1, dt:float=1e-5):
    """Generate the tone of the keypad given the number you push.
    
    We store the frequencies of the rows and columns and use the 
    equation in the paper to generate the tone.
    The tone is a wave function and we need the duration of the tone as well.
    The generated function is from time 0 to the requested duration.

    Parameters
    ----------
    num : int, optional
        The number for which the tone is produced, by default 0
    duration: float, optional
        The duration of the tone in [seconds], by default 1e-2
    dt: float, optional
        Time step for the time series in [seconds], by default 1e-5
    """
    frequency_row = [697, 770, 852, 941]
    frequency_col = [1209, 1336, 1477]
    
    if num == 0:
        row = 3 # fourth row
        col = 1 # middle column
    else:
        # for numbers from 1 to 9 we can use the pattern on the numpad
        # to find the row and column indices.
        row = (num - 1) % 3
        col = (num - 1) // 3
    # print(row, col)
    
    f_r = frequency_row[row]    # Frequency for the row
    f_c = frequency_col[col]    # Frequency for the column
    
    time = np.arange(0, duration, dt)
    tone = 0.5 * (np.sin(2 * np.pi * f_r * time) + np.sin(2 * np.pi * f_c * time))
    
    return time, dt, tone
```

```python
# Let's execute this method for num = 3
num = 9
time, dt, tone = gentone(num=num)

plt.plot(time, tone)
plt.grid(alpha=.5)
plt.show()
```

# Section (b)

```python
# We'll be using the numpy.fft.fft method to perform the FFT
fft = np.fft.fft(tone)
# We use numpy.fft.fftfreq to find the frequency bins for the FFT plot.
freqs = np.fft.fftfreq(tone.size, d=dt)

# Plot the real and imaginary components of the FFT
plt.plot(freqs, fft.real, 'c--', label="real")
plt.plot(freqs, fft.imag, 'r--', label="imag")
plt.xlim(-2000, 2000)
plt.grid(alpha=.5)
plt.xlabel("Freq [Hz]")
plt.ylabel("Power")
plt.title("FFT of the signal")
plt.legend()
plt.show()
```

## Section (c)

```python
def gen2tone(num1:int, num2:int):
    """Generate the two tone signal. First, num1 is pressed,
    and after 10 timesteps, num2 is pressed.

    Parameters
    ----------
    num1 : int
        First number to press
    num2 : int
        Second number to press
    """
    time1, dt1, tone1 = gentone(num1)
    time2, dt2, tone2 = gentone(num2, dt=dt1)
    n_steps = round(.05 // dt1)
    time_steps = time1.size + time2.size + n_steps
    duration = time_steps * dt1
    time = np.arange(0, duration, dt1)    
    middle = np.zeros((n_steps, ), dtype=float)
    signal = np.concatenate([tone1, middle, tone2])
    # print(signal.shape)
    # print(time.shape)
    signal += np.random.rand(signal.size) * 0.3    # add noise
    
    return time, dt1, signal
```

```python
time, dt, signal = gen2tone(6, 9)

plt.plot(time, signal)
plt.xlabel("time [s]")
plt.ylabel("intensity")
plt.title("2 number tone signal plot")
plt.grid(alpha=.5)
# plt.savefig("img/2tone_signal.jpg", bbox_inches='tight', dpi=200)
plt.show()
```

## Section (d)

```python
from scipy.signal import spectrogram

f, t, Sxx = spectrogram(signal, fs=1/dt)

print(f.shape, t.shape, Sxx.shape)
```

```python
plt.pcolormesh(t, f[:10], np.abs(Sxx[:10, :]), vmin=0)
plt.xlabel("Time [s]")
plt.ylabel("Freq [Hz]")
plt.title("STFT of the 2 tone signal")
plt.colorbar()
```
