```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load the Data

```python
data = pd.read_csv("cars-fixed04.dat")
print(data.size)

# Take only the numerical features of the data
numerical_cols = data.columns[7:]
data_num = data[numerical_cols]
data_num.head()
```

# Section (a)

```python
sns.pairplot(data=data_num, dropna=True, kind='scatter')
plt.show()
```

# Section (b)

```python
# First we need to standardize the dataset. So we substract by mean and divide by std
data_standard = (data_num - data_num.mean()) / data_num.std()

cov = data_standard.cov()
print(cov)
sns.heatmap(data=cov, cmap=plt.cm.plasma)
plt.show()
```

# Section (c)
For the PCA, I'm going to use the PCA module by sklearn.
First, I'll keep all the components to draw the scree plot. Then, I'll choose the number of components needed to compress it for section (d)

```python
from sklearn.decomposition import PCA

data_arr = data_standard.to_numpy()
pca = PCA() # by default, all components are kept.
pca.fit(data_arr)   # fit the data to the PCA
print(pca.explained_variance_ratio_)   # print the explained variance by every PC
```

```python
# Get the values for the scree plot
variance = pca.explained_variance_ratio_
scree = np.cumsum(variance)

plt.plot(np.arange(scree.size), scree, label="variance explained totally")
plt.plot(np.arange(scree.size), variance, label="variance for each PC")
plt.grid(alpha=.5)
plt.title("Scree plot for the PCA")
plt.xlabel("Number of PC used")
plt.ylabel("Percentage of Information explained")
plt.legend(loc='best')
plt.show()
```

We can see that using the first two PCs gives us more than 80% of the data.

# Section (d)

```python
pca2 = PCA(n_components=2)
data_reduced = pca2.fit_transform(data_arr)
print(pca2.explained_variance_ratio_)
print(data_reduced.shape)
```

```python
plt.scatter(data_reduced[:, 0], data_reduced[:, 1])
plt.grid(alpha=.5)
plt.show()
```
